#### Analysis of Civil Rights Distribution

```python
columns = df.columns[2:]
summary = df.groupby('year')[columns].agg(['median', 'std', 'min', 'max', lambda bottom_five: bottom_five.quantile(0.05), lambda top_95: top_95.quantile(0.95)])

metrics = ["median", "min", "max", "std"]

for metric in metrics:
    plt.figure(figsize=(10, 6))

    for indicator in columns:
        label_clean = indicator.replace("_mean", "").upper()
        ax = plt.gca()
        ax.plot(
            summary.index, 
            summary[(indicator, metric)], 
            marker='o', 
            label=label_clean
        )

    plt.title(f'Civil Rights Indicators: {metric.upper()} Over Time')
    plt.xlabel('Year')
    plt.ylabel(metric.capitalize())
    
    # Shade Trump (2017-2018) and Biden (2021-2022) start periods
    plt.axvspan(2017, 2019, color='orange', alpha=0.2, label='Trump Start')
    plt.axvspan(2021, 2023, color='blue', alpha=0.2, label='Biden Start')

    handles, labels = ax.get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), title='Indicator', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Median: We notice that the trend for the median country is similar to the trend for the mean of the countries. Furthermore, we notice that the median country experiences shifts more drastically than the global average would suggest. Ultimately, global civil rights protection worsened after 2020. 

Min: The worst countries' scores dipped sharply between 2017â€“2019 (early Trump years) and recovered a bit during 2021â€“2023 (Biden start).

Max: The best countries reached peak scores around 2019â€“2020. Then slight declines in maximum scores in recent years (2021â€“2023). This demonstrates that even the best countries experienced worsened civil-rights after 2020.

Standard Deviation: We notice that the standard deviation of indicators was relatively uncorrelated in 2017-2019. However, after 2020, the standard deviation of indicators began to decrease in a highly correlated fashion. This suggests that the civil-rights for the countries in the dataset began to converge in 2020, representing a foundational shift in human rights which brought the world closer together

From this long-term graph, we see differing long-term trends of the indicators. Because of this, we should calculate a line of best fit of each indicator up until the beginning of the analysis window, to be able to better analyze the impact of Trump and Obama on the global indictors

```python
from sklearn.linear_model import LinearRegression

# Filter data up until 2017
df_pre_2017 = df_yearly_mean[df_yearly_mean.index <= 2017]

best_fit_lines = {}

for column in df_pre_2017.columns:
    data = df_pre_2017[[column]].dropna()
    
    if not data.empty:
        X = data.index.values.reshape(-1, 1)
        y = data[column].values
        
        model = LinearRegression()
        model.fit(X, y)
        
        best_fit_lines[column] = {
            'slope': model.coef_[0],
            'intercept': model.intercept_
        }
```

We have a lot of indicators, so we will separate the data into positive-trends and negative-trends for cleaner analysis.

#### Best Fit Lines

```python
positive_slope = {key: value for key, value in best_fit_lines.items() if value['slope'] > 0}
negative_slope = {key: value for key, value in best_fit_lines.items() if value['slope'] < 0}
```

```python
# Filter data for post-2017
df_post_2017 = df_yearly_mean[df_yearly_mean.index > 2017]

def plot_trends(columns_dict, title):
    plt.figure(figsize=(12, 6))
    for i, column in enumerate(columns_dict.keys()):
        data = df_post_2017[[column]].dropna()
        if not data.empty:
            # Actual line
            plt.plot(data.index, data[column], label=f'{column} (Actual)', color=f'C{i}')
            
            # Best fit
            X = data.index.values.reshape(-1, 1)
            model = LinearRegression().fit(X, data[column].values)
            y_pred = model.predict(X)
            plt.plot(data.index, y_pred, linestyle='dotted', color=f'C{i}', label=f'{column} (Best Fit)')

    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel('Score')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Plot separately
plot_trends(positive_slope, "Post-2017 Indicators with Positive Trends")
plot_trends(negative_slope, "Post-2017 Indicators with Negative Trends")
```

The diference appears extremely small, however that could be due to errors or inconsistencies in scale. Let's standardize and calculate the difference between the predicted and actual values.

```python
percent_diff_by_indicator = {}

for column in df_post_2017.columns:
    data = df_post_2017[[column]].dropna()
    if not data.empty:
        X = data.index.values.reshape(-1, 1)
        y_actual = data[column].values

        model = LinearRegression().fit(X, y_actual)
        y_pred = model.predict(X)

        y_pred_safe = y_pred.copy()
        y_pred_safe[y_pred_safe == 0] = 1e-6

        percent_diff = (y_actual - y_pred_safe) / y_pred_safe * 100
        percent_diff_by_indicator[column] = pd.Series(percent_diff, index=data.index)

df_percent_diff = pd.DataFrame(percent_diff_by_indicator)

# Plot
ax = df_percent_diff.plot(marker='o', figsize=(12, 6))
plt.title("Percentage Difference from Best-Fit Trend (Post-2017)")
plt.xlabel("Year")
plt.ylabel("Percent Difference (%)")
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.legend(title="Indicator", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

From this analysis, we confirm that the majority of the economic prosperity indicators do not seem to be shifting relative to their historical trends. However, we should look at the standard deviation of these indicators to be certain that the administrations are not having significant impact.

However, we do note that the not-unemployed score and work score saw significant increase between 2018 and 2020. The quality of life score also saw significant increase post-2018.

#### Standard Dev

```python
std_diff_by_indicator = {}

for column in df_post_2017.columns:
    data = df_post_2017[[column]].dropna()
    if not data.empty:
        X = data.index.values.reshape(-1, 1)
        y_actual = data[column].values

        model = LinearRegression().fit(X, y_actual)
        y_pred = model.predict(X)

        residuals = y_actual - y_pred
        std_resid = residuals.std(ddof=1) if residuals.std(ddof=1) != 0 else 1e-6

        std_dev_diff = residuals / std_resid
        std_diff_by_indicator[column] = pd.Series(std_dev_diff, index=data.index)

df_std_diff = pd.DataFrame(std_diff_by_indicator)

# Plot
ax = df_std_diff.plot(marker='o', figsize=(12, 6))
plt.title("Standard Deviation from Best-Fit Line (Post-2017)")
plt.xlabel("Year")
plt.ylabel("Deviation from Trend (Standard Deviations)")
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.axhline(2, color='red', linestyle=':', linewidth=0.7)
plt.axhline(-2, color='red', linestyle=':', linewidth=0.7)
plt.legend(title="Indicator", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

Aha! For almost every indicator and year, the deviation from the trend is within Â±1 standard deviation. This suggests that the observed differences are small and likely due to random variation rather than meaningful departures from trend due to external factors.
